```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_summary = pd.DataFrame(school_data_df)
school_data_summary.head()
```

## District Summary

```python
# Calculate the total number of unique schools
unique_school_name = school_data_summary['school_name'].unique()
total_school_number = len(unique_school_name)
total_school_number
```

```python
# Calculate the total number of students
student_count = school_data_summary['Student ID'].count()
student_count
```

```python
# Calculate the total budget
total_budget = school_data['budget'].sum()
#Had to explicitly declare as 'float' since DataFrame would not recognize
total_budget.astype(float)
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_summary['math_score'].mean(numeric_only = True)
#Had to explicitly declare as 'float' since DataFrame would not recognize
average_math_score.astype(float)
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_summary['reading_score'].mean(numeric_only = True)
#Had to explicitly declare as 'float' since DataFrame would not recognize
average_reading_score.astype(float)
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_summary[(school_data_summary["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_summary[(school_data_summary['reading_score']>= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count/float(student_count) * 100
passing_reading_percentage.astype(float)
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_summary[
    (school_data_summary["math_score"] >= 70) & (school_data_summary["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
#Sets overall_passing_rate
overall_passing_rate.astype(float)
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
                                "Total Schools":total_school_number,
                                 "Total Students":f"{student_count:,}",
                                 "Total Budget":f"${total_budget:,.2f}",
                                 "Average Math Score":f"{average_math_score:.6f}",
                                 "Average Reading Score":f"{average_reading_score:.5f}",
                                 "% Passing Math":f"{passing_math_percentage:.6f}",
                                 "% Passing Reading":f"{passing_reading_percentage:.6f}",
                                 "% Overall Passing":f"{overall_passing_rate:.6f}"
                                 },index=[0])

# Formatting
#district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
#district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_summary[
    (school_data_summary["reading_score"] >= 70) & (school_data_summary["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).count()['student_name']/per_school_counts*100
school_students_passing_math_and_reading
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_summary[school_data_summary["grade"] == "9th"]
tenth_graders = school_data_summary[(school_data_summary["grade"] == "10th")]
eleventh_graders = school_data_summary[(school_data_summary["grade"] == "11th")]
twelfth_graders = school_data_summary[(school_data_summary["grade"] == "12th")]

#ninth_graders.head()

# Group by `school_name` and take the mean of the `math_score` column for each.
#Be sure to include (numeric_only=True), will not work since there are strings
ninth_grade_average = ninth_graders.groupby(['school_name']).mean(numeric_only=True)['math_score']
tenth_grade_average = tenth_graders.groupby(['school_name']).mean(numeric_only=True)['math_score']
eleventh_grade_average = eleventh_graders.groupby(['school_name']).mean(numeric_only=True)['math_score']
twelfth_grade_average = twelfth_graders.groupby(['school_name']).mean(numeric_only=True)['math_score']



# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({
"School Names": schoolnames,
"Ninth Graders Math Scores": ninth_grade_average,
"Tenth Graders Math Scores": tenth_grade_average,
"Eleventh Graders Math Scores": eleventh_grade_average,
"Twelfth Graders Math Scores": twelfth_grade_average})

#Code to format Scores to 1 decimal place

math_scores_by_grade["Ninth Graders Math Scores"] = math_scores_by_grade["Ninth Graders Math Scores"].map("{:.1f}".format)
math_scores_by_grade["Tenth Graders Math Scores"] = math_scores_by_grade["Tenth Graders Math Scores"].map("{:.1f}".format)
math_scores_by_grade["Eleventh Graders Math Scores"] = math_scores_by_grade["Eleventh Graders Math Scores"].map("{:.1f}".format)
math_scores_by_grade["Twelfth Graders Math Scores"] = math_scores_by_grade["Twelfth Graders Math Scores"].map("{:.1f}".format)

math_scores_by_grade = math_scores_by_grade.reset_index(drop=True)

# Minor data wrangling
#math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
#ninth_grade
#ninth_grade_average
#ninth_graders
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
#school_spending_df = school_data_summary.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_capita,spending_bins,labels=labels,ordered = True)
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(['Spending Ranges (Per Student)'])['Average Math Score'].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean(numeric_only=True)
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean(numeric_only=True)
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean(numeric_only=True)
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean(numeric_only=True)

#Display Values

#spending_math_scores
#spending_reading_scores
#spending_passing_math
#spending_passing_reading
overall_passing_spending
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({
    #"Spending Ranges (Per Student)":labels,
    "Average Math Score":spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math":spending_passing_math,
    "% Passing Reading":spending_passing_reading,
    "% Overall Passing": overall_passing_spending
    
})

# Display results
spending_summary
```
